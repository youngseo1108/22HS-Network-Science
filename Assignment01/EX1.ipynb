# A01.2a Global properties

```python
g_k = nx.read_gml('data/graph_karate.gml')
g_n = nx.read_gml('data/graph_netsci.gml')
g_f = nx.read_gml("data/graph_facebook.gml")

# The degree of each network
g_kdegrees = g_k.degree()
g_ndegrees = g_n.degree()
g_fdegrees = g_f.degree()

def average_degree(degrees):
    sum = 0
    for node in degrees:
        sum = node[1] + sum
    return sum/len(degrees)

print('Average degree Karate:')
print(average_degree(g_kdegrees))
print('Average degree Netsci:')
print(average_degree(g_ndegrees))
print('Average degree Facebook:')
print(average_degree(g_fdegrees))
```

```python
# The density of each network
g_kdensity = nx.density(g_k)
g_ndensity = nx.density(g_n)
g_fdensity = nx.density(g_f)

print('Density Karate:')
print(g_kdensity)
print('Density Netsci:')
print(g_ndensity)
print('Density Facebook:')
print(g_fdensity)
```

```python
''' Write a function max_degree that takes a network as its argument, and returns two values: 
the id/name/label of the node with the largest degree, and the value of its degree. Test it on the datasets given'''

def max_degree(network):
    # get a degree of a network
    net_dic = dict(network.degree())
    # set the key & value of the first element as max
    max_label = list(net_dic.keys())[0]
    max_degree = net_dic[max_label]
    # itereate throughout the dictionary of degrees(value) & nodes(key)
    for k in net_dic.keys():
        if max_degree < net_dic[k]:
            max_degree = net_dic[k]
            max_label = k
    return (max_label, max_degree)

print('Max degree Karate:')
print(max_degree(g_k))
print('Max degree Netsci:')
print(max_degree(g_n))
print('Max degree Facebook:')
print(max_degree(g_f))
```
